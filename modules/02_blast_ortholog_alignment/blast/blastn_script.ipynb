### List and Parameters
#### E-Value
- 0.05 	- Default in NCBI Blast (https://blast.ncbi.nlm.nih.gov/Blast.cgi) 
- 1e-10 - Default in-website blast in solgenomics (https://solgenomics.net/tools/blast/?db_id=320).

#### Word Size
- smaller word size, more sensitive, but increases false positive; 
- longer word size, less sensitive, but require specificity (longer exact matches)
- 11 - Default in solgenomics (https://solgenomics.net/tools/blast/?db_id=320).

```python
%%bash

# Databases (Date Accessed: September 22, 2024)
Eggplant_V3_Chromosomes_DB="$assembly_chr_FASTA_DIR/Eggplant_V3_Chromosomes_db/Eggplant_V3_Chromosomes"
Eggplant_V3_CDS_putative_function_DB="$assembly_chr_FASTA_DIR/Eggplant_V3_CDS_putative_function_db/Eggplant_V3_CDS_putative_function"

# DMPs from Arabidopsis thaliana, Oryza sativa, and Zea mays (Date Accessed: September 22-26, 2024)

# AtDMPs from (https://www.arabidopsis.org/).
AtDMPs_fasta_FILE=/mnt/e/_ELBI/_MASTER/MBB_300_Master_Thesis/_DMP_Eggplant/01_Identification/query_fasta_file/Arabidopsis_thaliana/v1_keyword_searched/AtDMPs_nuc_seq_merged_v1.fasta
AtDMPs_fasta_FILE=/mnt/e/_ELBI/_MASTER/MBB_300_Master_Thesis/_DMP_Eggplant/01_Identification/query_fasta_file/Arabidopsis_thaliana/v2_based_from_paper/AtDMPs_nuc_seq_merged_v2.fasta

# OsDMPs from the query DMP sequences use in the paper of <paper SWEETPOTATO>, and sequences lifted from annotated sequences obtained from (https://rice.uga.edu/).
OsDMPs_fasta_FILE=/mnt/e/_ELBI/_MASTER/MBB_300_Master_Thesis/_DMP_Eggplant/01_Identification/query_fasta_file/Oryza_sativa/OsDMPs_nuc_seq_merged.fasta

# ZmDMPs from a simple keyword search of "DMP" in this website, (https://www.maizegdb.org/gene_center/gene).
ZmDMPs_v1_fasta_FILE=/mnt/e/_ELBI/_MASTER/MBB_300_Master_Thesis/_DMP_Eggplant/01_Identification/query_fasta_file/Zea_mays/v1/ZmDMPs_nuc_seq_merged_v1.fasta
ZmDMPs_v2_fasta_FILE=/mnt/e/_ELBI/_MASTER/MBB_300_Master_Thesis/_DMP_Eggplant/01_Identification/query_fasta_file/Zea_mays/v2/ZmDMPs_nuc_seq_merged_v2.fasta

# List and Parameters
Eggplant_db_LIST=($Eggplant_V3_Chromosomes_DB $Eggplant_V3_CDS_putative_function_DB)
DMP_fasta_LIST=($OsDMPs_fasta_FILE $AtDMPs_fasta_FILE $ZmDMPs_v1_fasta_FILE $ZmDMPs_v2_fasta_FILE)
e_value_LIST=(0.05 1e-10)
word_size_LIST=(11 20)

for DMP_fasta_FILE in ${DMP_fasta_LIST[@]}; do 
	for DB in ${Eggplant_db_LIST[@]}; do
		for e_value in ${e_value_LIST[@]}; do
			for word_size in ${word_size_LIST[@]}; do
				DMP_fasta_BASENAME=$(basename "$DMP_fasta_FILE" .fasta)
				DB_BASENAME=$(basename "$DB")
				echo "RUN3"
				mkdir -p "01_Identification2/blastn_results/ev_${e_value}_ws_${word_size}"	
				blastn_result_csv_VAR="01_Identification2/blastn_results/ev_${e_value}_ws_${word_size}/blastn_${DB_BASENAME}_VS_${DMP_fasta_BASENAME}_${e_value}_${word_size}.csv"
				blastn_result_txt_VAR="01_Identification2/blastn_results/ev_${e_value}_ws_${word_size}/blastn_${DB_BASENAME}_VS_${DMP_fasta_BASENAME}_${e_value}_${word_size}.txt"

				echo "$DMP_fasta_BASENAME is being aligned with $DB_BASENAME." 
				echo "Query ID,Subject ID,Percent Identity,Alignment Length,Mismatches,Gaps,E-value,Bit Score" > "$blastn_result_csv_VAR" #cat - $result_file_VAR > temp.csv && mv temp.csv $result_file_VAR
				blastn -db "$DB" \
					-query $DMP_fasta_FILE \
					-evalue $e_value \
					-word_size $word_size \
					-outfmt "10 qseqid sseqid pident length mismatch gaps evalue bitscore" >> "$blastn_result_csv_VAR"

				blastn -db "$DB" \
					-query $DMP_fasta_FILE \
					-evalue $e_value \
					-word_size $word_size > $blastn_result_txt_VAR

				echo -e "Alignment of $DMP_fasta_BASENAME with $DB_BASENAME is a success.\n"
			done
		done
	done
done

```
